## 0) Setup

```python
# Run it first.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins.head()  # preview
```

## 1–3) Scatter plot with color by species
Create a scatter plot with:
- **x =** `bill_length_mm`
- **y =** `bill_depth_mm`
- Color points by **species**

Then add a short comment (2–3 sentences) describing the pattern(s) you see.

```python
# TODO: Create the scatter plot here

# Scatter plot: bill_length vs bill_depth, colored by species
plt.figure(figsize=(7,5))
sns.scatterplot(
    data=penguins,
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='species',
    style='species'
)
plt.title("Bill Length vs Bill Depth by Penguin Species")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.show()
```

**Your interpretation (2–3 sentences):**

_Write your observations here._

Gentoo penguins have bills that are longer and not as deep, which makes them stand out from the rest.  Adelie penguins have shorter bills and deeper bodies, and they are very close together.  Chinstrap penguins are in the middle of the two groups, making a clear mid-range pattern.

## 4) Histogram with two different bin sizes
Plot a histogram of `bill_length_mm` **twice**:
- Once with **10 bins**
- Once with **30 bins**

Then write 1–2 sentences comparing what changes when you change the number of bins.

```python
# TODO: Histogram with 10 bins

# Histogram with 10 bins
plt.figure(figsize=(7,4))
sns.histplot(penguins['bill_length_mm'], bins=10)
plt.title("Histogram of Bill Length (10 bins)")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Count")
plt.show()
```

```python
# TODO: Histogram with 30 bins

# Histogram with 30 bins
plt.figure(figsize=(7,4))
sns.histplot(penguins['bill_length_mm'], bins=30)
plt.title("Histogram of Bill Length (30 bins)")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Count")
plt.show()
```

**Your comparison (1–2 sentences):**

_Write your observations here._


The distribution seems smooth and wide with 10 bins, which gives a basic idea of the range of bill lengths.  With 30 bins, more detailed peaks and valleys show up, showing minor patterns that a broader bin setting hides.


## 5) Boxplot of body mass by species
Create a **boxplot** showing `body_mass_g` grouped by `species`.
Then write 1–2 sentences summarizing which species are heavier/lighter on average.

```python
# TODO: Boxplot of body mass by species

# Boxplot of body mass by species
plt.figure(figsize=(7,5))
sns.boxplot(data=penguins, x='species', y='body_mass_g')
plt.title("Body Mass by Penguin Species")
plt.xlabel("Species")
plt.ylabel("Body Mass (g)")
plt.show()

```

**Your summary (1–2 sentences):**

_Write your observations here._

The heaviest penguins are Gentoo penguins, and the lightest are Adelie penguins.  Chinstrap penguins are in the middle, with body mass levels that are average.
